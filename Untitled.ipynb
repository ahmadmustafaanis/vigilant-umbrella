```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Real estate.csv')
```

```python
df.head()
```

```python
df.info()
```

### Splitting data set in data and targets

```python
X = df.iloc[:,2:-1]
X.head()
```

```python
y = df.iloc[:,-1]
y.head()
```

### Splitting data set in training and testing data

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=1)
```

```python
print(f"Training Data Shape{X_train.shape}")
print(f"Training Labels Shape{y_train.shape}")
print(f"Testing Data Shape{X_test.shape}")
print(f"Testing Labels Shape{y_test.shape}")
```

### Checking Variance of Data Set and then Standardizing it

```python
from sklearn.preprocessing import StandardScaler
print(df.var())
```

```python
ss = StandardScaler()
ss.fit(X_train)
X_test_std = ss.transform(X_test)
X_train_std = ss.transform(X_train)
```

```python
X_train_std.var(), X_train_std.mean()
```

```python
X_test_std.var(), X_test_std.mean()
```

### Making a Linear Regression Model with Random Forest Machine Learning Algorithm

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model1 = RandomForestRegressor(warm_start=True, oob_score=True)
```

```python
model1.fit(X_train_std, y_train)
```

```python
predictions = model1.predict(X_test_std)
```

```python
predictions
```

```python
y_test.to_numpy()
```

#### If we analyze our predictions and and `y_test` we have done a good job in predicting values, Let's now Use some metrics to evaluate our Model

```python
from sklearn.metrics import max_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import classification_report
```

```python
print(max_error(y_test, predictions))
```

```python
print(mean_absolute_error(y_test, predictions))
```

```python
print(mean_squared_error(y_test, predictions))
```

```python
print(r2_score(y_test, predictions))
```

### Let's Plot the Regression Line

```python
sns.regplot(y_test,predictions)
plt.scatter(y_test, predictions, color=['red','green']*41 + ['red']);
```

#### Here we Can see that Our Predicted Values(`Red`) are very much close to Actuall Values(`Green`) and Regression Line is fitting good.

#### After Analyzing these Metrics we can say that our model is performing pretty well as `mean_absoulte_error`, `r2_score` and `max_error` suggest.
#### Let's Use some Deep Learning Now to Improve this scores further

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
model2 = Sequential([
    Dense(512,input_shape=X_train_std[0].shape, activation='relu', kernel_regularizer='l2'),
    Dense(256, activation='relu'),
    Dropout(0.3),
    Dense(32, activation='relu', kernel_regularizer='l1'),
    Dense(1)
])

print(model2.summary())
```

```python
model2.compile(optimizer='adam', loss='mse', metrics=['mse', 'mae'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping

es = EarlyStopping(patience=5)
```

```python
hist = model2.fit(X_train_std, y_train.to_numpy(), validation_data = (X_test_std, y_test.to_numpy()), epochs=30,callbacks=[es] )
```

### Let's Plot to get more intuition

```python
plt.plot(hist.history['loss'], label="Loss")
plt.plot(hist.history['val_loss'], label="Validaton Loss")
plt.title("Loss vs Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend();
```

```python
plt.plot(hist.history['mse'], label="MSE")
plt.plot(hist.history['val_mse'], label="Validaton MSE")
plt.title("MSE vs Validation MSE")
plt.xlabel("Epochs")
plt.ylabel("MSE")
plt.legend();
```

```python
plt.plot(hist.history['mae'], label="MAE")
plt.plot(hist.history['val_mae'], label="Validaton MAE")
plt.title("MAE vs Validation MAE")
plt.xlabel("Epochs")
plt.ylabel("MAE")
plt.legend();
```

```python
pred2 = model2.predict(X_test_std).T
pred2
```

### Let's Plot the Regression Line

```python
sns.regplot(y_test,pred2)
plt.scatter(y_test, pred2, color=['red','green']*41 + ['red']);
```

```python
model2.evaluate(X_test_std, y_test.to_numpy(), verbose=2)
```

### In Deep Learning our MSE is `43`, where as in Random Forest it was `64` and MAE is 5.2 as compared to 5.3
